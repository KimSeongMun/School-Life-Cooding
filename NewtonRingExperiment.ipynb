```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

i = input("Dataset을 고르세요: ")
j = input("기울기 값을 고르세요: ")
if i == 'Yello':
    x1 = np.array([0, 18.3, 36.6, 54.9, 73.2])
    x2 = np.array([0, 0.000000578, 0.000001156, 0.000001734, 0.000002312])
    y = np.array([0.00315**2, 0.00365**2, 0.00555**2, 0.00645**2, 0.00715**2])
    X = np.linspace(min(x1), max(x1), 5)
elif i == 'Green':
    x1 = np.array([0, 18.3, 36.6, 54.9, 73.2])
    x2 = np.array([0, 0.000000546, 0.000001092, 0.000001638, 0.000002184])
    y = np.array([0.0031 ** 2, 0.0042 ** 2, 0.00525 ** 2, 0.00605 ** 2, 0.007 ** 2])
    X = np.linspace(min(x1), max(x1), 5)
elif i == 'Blue':
    x1 = np.array([18.3, 36.6, 54.9])
    x2 = np.array([0.000000436, 0.000000872, 0.000001308])
    y = np.array([0.00315 ** 2, 0.00415 ** 2, 0.0051 ** 2])
    X = np.linspace(min(x1), max(x1), 3)
else:
    print("Dataset을 잘못 선택하였습니다.")
if j == 'lambda':
    model = LinearRegression()
    model.fit(x1.reshape(-1,1), y.reshape(-1,1))
    R2_score = model.score(x1.reshape(-1,1), y.reshape(-1,1))
    coefficient = float(model.coef_)
    intercept = float(model.intercept_)
    print(" R2_score : {0} \n Coefficient : {1} \n Intercept : {2}".format(R2_score, coefficient, intercept))
    Y = coefficient * X + intercept

    plt.scatter(x1, y, alpha = 0.2, s = 20, c = 'b')
    plt.plot(X, Y, 'c', linewidth=1 ,label=" R2_score : {0} \n Coefficient : {1} \n Intercept : {2}".format(R2_score, coefficient, intercept))
    plt.xlabel('(N-1)R,m')
    plt.ylabel('r_n_square,m_square')
    plt.grid()
    plt.legend()
    if i == 0:
        plt.title('Yello')
    elif i == 1:
        plt.title('Green')
    else:
        plt.title("Blue")
elif j == 'R':
    model = LinearRegression()
    model.fit(x2.reshape(-1, 1), y.reshape(-1, 1))
    R2_score = model.score(x2.reshape(-1, 1), y.reshape(-1, 1))
    coefficient = float(model.coef_)
    intercept = float(model.intercept_)
    print(" R2_score : {0} \n Coefficient : {1} \n Intercept : {2}".format(R2_score, coefficient, intercept))
    Y = coefficient * X + intercept

    plt.scatter(x2, y, alpha=0.2, s=20, c='b')
    plt.plot(X, Y, 'c', linewidth=1,
             label=" R2_score : {0} \n Coefficient : {1} \n Intercept : {2}".format(R2_score, coefficient, intercept))
    plt.xlabel('(N-1)lambda,m')
    plt.ylabel('r_n_square,m_square')
    plt.grid()
    plt.legend()
    if i == 0:
        plt.title('Yello')
    elif i == 1:
        plt.title('Green')
    else:
        plt.title("Blue")
plt.show()
```
